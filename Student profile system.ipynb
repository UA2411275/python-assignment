```python
print ("Enter your name")
name=input()
print ("You entered:",name)
```

```python
print ("Enter your age")
age=input()
age=int(age)
```

```python
print ("Enter your GPA")
gpa=input()
gpa=float(gpa)
```

```python
print ("Enter courses")
courses=input()
courses=courses.split(",")
courses=[c.strip() for c in courses]
```

```python
student_profile={
    "name":name,
    "age":age,
    "gpa":gpa,
    "courses":courses,
    "num_courses":len(courses)
}
```

```python
is_dean=gpa>=3.5
age_status="Adult" if age>=18 else "minor"
```

```python
print ("---Types conversion---\n")
print ("Age is now:",age,"|Type:",type(age))
print ("GPA is now:",gpa,"|Type:",type(gpa))
```

```python
print ("\n=========STUDENT PROFILE===========")
print(f"Name:{student_profile['name']}")
print(f"Age:{student_profile['age']} ({age_status})")
print(f"Gpa:{student_profile['gpa']:.2f}")
print(f"Courses:{student_profile['courses']}")
print(f"#Courses:{student_profile['num_courses']}")
print(f"Dean's List:{'Yes' if is_dean else'no'}")
print("====================")
```
